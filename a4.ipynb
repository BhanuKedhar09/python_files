```python
import os
import string
import numpy as np
from collections import Counter
from urllib.request import urlretrieve
from scipy.stats import chi2_contingency

def download_text(url):
    local_fname = os.path.basename(url)
    if not os.path.exists(local_fname):
        urlretrieve(url, local_fname)
    return local_fname
```

```python

def read_data(file_name):
    global book_name
    all_lines = open(file_name, 'r', encoding='utf-8-sig')
    li = []
    for line in all_lines:
        if '*** START' in line:
            book_name = line.strip() 
            for ano in all_lines:
                if '*** END' in  ano:
                    break
                if ano.strip() != '':
                    li.append(ano.strip())
            break
    return li 

book_list = read_data(fname)  # 3363
book_list
```

<h3>Q.2) Find Non-ASCII Characters </h3>

```python
def find_non_ascii(send):
    s = set()
    for li in send:
        for l in li:
            if ord(l) > 127:
                s.add(l)
            else:
                continue
    return(s)

send = read_data(fname)
s1 = find_non_ascii(send)
```

```python

def diacriticss(lis1):
    dicto = {"ä" : "ae", "ö" : "oe", "ü" : "ue","ß" : "ss"}
    dicto_upper = dict()
    for k in dicto:
        if k != 'ß':
            dicto_upper[k.upper()] = dicto[k].capitalize()
    dicto.update(dicto_upper)
    new_list1= []
    for line in lis1:
        lina = ""
        for char in line:
            if char in dicto.keys():
                # lina = line.replace(char, dicto[char])
                lina = lina + dicto[char]
            else :
                lina = lina + char
            
        new_list1.append("".join(lina))
    return new_list1
            
            
diacriticss(send)
```

<h3>Q.4) Compute Word Frequencies</h3>

```python
import string
from collections import Counter
def Compute_Word_Frequencies(g):
    res_lis = []
    for line in g:
        for cha in line:
            if cha in string.punctuation:
                line = line.replace(cha, '')
        res_lis.append(line)
    ss = set()
    word_frequency = Counter()
    for line in res_lis:
        inlower = []
        words = line.split()
        for word in words :
            ss.add(word)
            inlower.append(word.lower())
        word_frequency.update(Counter(inlower))
        
    infrequently = 0
    frequently = 0
    for i in word_frequency:
        if word_frequency[i] > 4 :
            frequently += 1
        else :
            infrequently += 1
    return [int(infrequently), int(frequently), int(len(word_frequency))]
g = diacriticss(send)
result_4 = Compute_Word_Frequencies(g)
```
